## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resizing the graph 
plt.figure(figsize=(5,3), dpi=110)

#plt.plot(x,y, label='2x', color='4169E1', linewidth=2, linestyle='--', marker='.', markersize=8, markeredgecolor='black', markerfacecolor='yellow')
#Instead of this I'm gonna go with SHORT HAND NOTATION
#ftm = '[color][marker][line]'
plt.plot(x,y, 'b.--', label='2x')

#Labelling the graph
plt.title('Our First Graph!', fontdict={'fontname':'DejaVu Sans', 'fontsize':15})
plt.xlabel('x axis', fontdict={'fontname':'DejaVu Serif', 'fontsize':8})
plt.ylabel('y axis', fontdict={'fontname':'DejaVu Serif', 'fontsize':8})

#Line Number Two
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2,'r--')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

## Bar Graph

```python
labels=['A', 'B', 'C']
values=[2,5,3.5]

#Plotting a bar graph
bars = plt.bar(labels, values)

#Selecting patterns for my data
#bars[0].set_hatch('o')
#bars[1].set_hatch('/')
#bars[2].set_hatch('*')

#This can also be used
patterns=['o', '/', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())

#Setting the figure size
plt.figure(figsize=(6, 4), dpi=300)
plt.savefig('barGraph.png', dpi=300)
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time(in USD)', fontdict={'fontname':'DejaVu Serif', 'fontsize':15})

plt.xlabel('Year', fontdict={'fontname':'DejaVu Sans', 'fontsize':10})
plt.ylabel('Price (in USD)', fontdict={'fontname':'DejaVu Sans', 'fontsize':10})

plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')

print(gas.Year[::3])

plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```

## Load Necessary Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

## Histograms

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#4169E1')

plt.xticks(bins)

plt.ylabel('Number of player')
plt.xlabel('Skill level')
plt.title('Distribution of player skills', fontdict={'fontname':'DejaVu Sans Mono'})

plt.show()
```

## Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count().iloc[0]

labels = ['Left', 'Right']
colors = ['#ed8505', '#7b74e3']
plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%')

plt.legend(labels) 

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')
plt.figure(figsize=(8,5))

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa.loc[(fifa.Weight >=175 & (fifa.Weight < 200))].count().iloc[0]
heavy = fifa.loc[fifa.Weight >= 200].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [.4,.2,0,0,.4]

plt.title('Weight Distribution of Fifa Players(in lbs)', fontdict = {'fontname':'DejaVu Serif'})
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.5, explode=explode)
plt.legend(labels)
plt.show()
```

## Box and Whiskers

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['Fc Barcelona', 'Real Madrid', 'New England Revolution']
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    box.set(color='#4169E1', linewidth=2)
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison', fontdict={'fontname':'DejaVu Serif'})
plt.ylabel('Overall Rating')

plt.show()
```
